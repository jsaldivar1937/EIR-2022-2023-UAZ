```python
import sys 
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image, ImageOps
%matplotlib inline
```

```python
#img = plt.imread(r'C:\Users\Usuario\Documents\Python\bote.jpg')
img = plt.imread(r'C:\Users\Usuario\Documents\Python\im_contrast_2.tiff')
fig1=plt.figure(figsize = (8,8))
plt.imshow(img, cmap='gray',vmin=0,vmax=255)
plt.axis('off');
```

```python
# img.shape
fig2 = plt.figure(figsize = (15,8))
a = fig2.add_subplot(1, 2, 1)
plt.imshow(img, cmap='gray',vmin=0,vmax=255)
plt.axis('off');
a.set_title('Original Image')
inmax = np.amax(img)
inmin = np.amin(img)
b = fig2.add_subplot(1, 2, 2)
b.set_title('Histogram')
arr = np.array(img)
arr = arr.ravel()
b, bins, patches = plt.hist(arr, 255)
plt.xlim([0,255])
plt.show()
```

```python
Lout=0;
Hout=255;
m=(Hout-Lout)/(inmax-inmin);
b=Lout-m*inmin;
Iout=img*m+b;

fig3 = plt.figure(figsize = (15,8))
a = fig3.add_subplot(1, 2, 1)
plt.imshow(Iout, cmap='gray',vmin=0,vmax=255)
plt.axis('off');
a.set_title('Improvement Image')

b = fig3.add_subplot(1, 2, 2)
b.set_title('Histogram')
arr = np.array(Iout)
arr = arr.ravel()
b, bins, patches = plt.hist(arr, 255)
plt.xlim([0,255])
plt.show()
```
