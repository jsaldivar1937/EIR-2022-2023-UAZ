```python
import sys 
import math 
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image, ImageOps
import cv2
%matplotlib inline
```

```python
img = cv2.imread(r'', cv2.IMREAD_GRAYSCALE)

fig1=plt.figure(figsize = (8,8))
plt.imshow(img, cmap='gray',vmin=0,vmax=255)
plt.axis('off');
print(img.shape)
```

```python
gamma=0.3
img=img.astype('float')
inmax = np.amax(img)
Iout=img/inmax
Iout=Iout**gamma
Iout=Iout*255


img=img.astype(np.uint8)
fig3 = plt.figure(figsize = (15,8))
a = fig3.add_subplot(1, 2, 1)
plt.imshow(img, cmap='gray',vmin=0,vmax=255)
plt.axis('off');
a.set_title('Original Image')

Iout=Iout.astype(np.uint8)
b = fig3.add_subplot(1, 2, 2)
plt.imshow(Iout, cmap='gray',vmin=0,vmax=255)
plt.axis('off');
b.set_title('Improvement Image')
```

```python
# img.shape
fig2 = plt.figure(figsize = (15,8))
a = fig2.add_subplot(1, 2, 1)
arr = np.array(img)
arr = arr.ravel()
plt.hist(arr, 255)
plt.xlim([0,255])


b = fig2.add_subplot(1, 2, 2)
arr = np.array(Iout)
arr = arr.ravel()
plt.hist(arr, 255)
plt.xlim([0,255])


```
